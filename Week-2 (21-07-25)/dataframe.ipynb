```python
# 1. Consider a list datatype (1D) then reshape it into2D, 3D matrix using numpy

import numpy as np

# Create a 1D list
my_list = [1, 2, 3, 4, 5, 6]

# Convert the list to a NumPy array
my_array = np.array(my_list)

try:
    my_2d_matrix = my_array.reshape(2, 3)
    print("2D Matrix:")
    print(my_2d_matrix)
except ValueError as e:
    print(f"Could not reshape to 2x3: {e}")

# Reshape the 1D array into a 3D matrix (e.g., 1 "layer", 2 rows, 3 columns)
try:
    my_3d_matrix = my_array.reshape(1, 2, 3)
    print("\n3D Matrix:")
    print(my_3d_matrix)
except ValueError as e:
    print(f"Could not reshape to 1x2x3: {e}")

# Example of reshaping that would fail due to incompatible dimensions
try:
    my_invalid_reshape = my_array.reshape(2, 4)
except ValueError as e:
    print(f"\nAttempting incompatible reshape failed as expected: {e}")
```

```python
# 2. Generate random matrices using numpy
import numpy as np

random_matrix_2d = np.random.rand(3, 4)
print("\nRandom 2D Matrix:")
print(random_matrix_2d)

random_matrix_3d = np.random.rand(2, 3, 5)
print("\nRandom 3D Matrix:")
random_matrix_3d
```

```python
# 3. Find the determinant of a matrix using scipy

from scipy import linalg

# For calculating determinant, the matrix must be square
square_matrix = np.random.rand(3, 3)
print("\nSquare Matrix for Determinant:")
print(square_matrix)

determinant = linalg.det(square_matrix)
print("\nDeterminant of the Square Matrix:")
determinant
```

```python
# 4. Find eigen value and eigen vector of a matrix using scipy

import numpy as np
from scipy import linalg

eigen_matrix = np.array([[1, 2], [3, 4]])
print("\nSquare Matrix for Eigenvalues and Eigenvectors:")
print(eigen_matrix)

eigenvalues, eigenvectors = linalg.eig(eigen_matrix)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
eigenvectors
```
