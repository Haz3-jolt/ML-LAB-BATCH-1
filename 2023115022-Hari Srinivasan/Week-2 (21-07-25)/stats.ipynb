Descriptive Statistics

Aim of the Experiment
The main aim of this experiment is to explore the given dataset. A sample database is created and is available in the file sample.csv.
                                                      Sample Dataset


The objectives of this experiment are:
1.       Explore all the statistical operations of Pandas and given in Listing 1
2.       Use Describe command and explore the dataset as given in Listing 2
3.       Use Descriptive Statistics for univariate and bivariate data as given in Listing 3


```python
# Descriptive Statistics

import pandas as pd

# Load the sample dataset
try:
  df_sample = pd.read_csv('dataset/sample.csv')
  print("\nSample dataset loaded:")
  print(df_sample)
except FileNotFoundError:
  print("\nsample.csv not found. Please make sure the file is created.")

# 1. Explore all the statistical operations of Pandas (Listing 1 equivalent - various stats)
print("\nExploring statistical operations on 'Marks' column:")
print(f"Count: {df_sample['Marks'].count()}")
print(f"Sum: {df_sample['Marks'].sum()}")
print(f"Mean: {df_sample['Marks'].mean()}")
print(f"Median: {df_sample['Marks'].median()}")
print(f"Mode: {df_sample['Marks'].mode().tolist()}")
print(f"Standard Deviation: {df_sample['Marks'].std()}")
print(f"Variance: {df_sample['Marks'].var()}")
print(f"Min: {df_sample['Marks'].min()}")
print(f"Max: {df_sample['Marks'].max()}")
print(f"Absolute: {df_sample['Marks'].abs().head()}")
print(f"Product: {df_sample['Marks'].prod()}") # May be large
print(f"Cumulative Sum:\n{df_sample['Marks'].cumsum().head()}")
print(f"Cumulative Product:\n{df_sample['Marks'].cumprod().head()}")
print(f"Rank:\n{df_sample['Marks'].rank().head()}")
print(f"Quantile (0.25): {df_sample['Marks'].quantile(0.25)}")

# 2. Use Describe command and explore the dataset (Listing 2 equivalent)
print("\nDescriptive statistics using .describe():")
print(df_sample.describe())

# 3. Use Descriptive Statistics for univariate and bivariate data (Listing 3 equivalent)

# Univariate Descriptive Statistics
print("\nUnivariate Descriptive Statistics for 'Marks':")
print(df_sample['Marks'].describe())

print("\nUnivariate Descriptive Statistics for 'gender':")
print(df_sample['gender'].describe())

# Bivariate Descriptive Statistics (e.g., Marks by Gender)
print("\nBivariate Descriptive Statistics (Marks by Gender):")
print(df_sample.groupby('gender')['Marks'].describe())

# Correlation between numerical columns (if more than one)
# In this simple dataset, only 'Marks' is numerical, so correlation isn't meaningful
# If there were more numerical columns, you could use df_sample.corr()

# Example of correlation with dummy numerical data
dummy_df = pd.DataFrame({
    'col1': [1, 2, 3, 4, 5],
    'col2': [5, 4, 3, 2, 1],
    'col3': [10, 8, 6, 4, 2]
})
print("\nCorrelation matrix for dummy data:")
print(dummy_df.corr())
```
